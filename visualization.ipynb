```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('recorder/Pong-v5.csv')
df.head()
```

```python

# Plot return over episodes.
plt.figure(figsize=(10, 6))
plt.plot(df['episode'], df['return'])
plt.xlabel('Episode')
plt.ylabel('Return')
plt.title('Pong-v5 Training')
plt.tight_layout()
```
